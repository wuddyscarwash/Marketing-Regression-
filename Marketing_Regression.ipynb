# Simple Linear Regression: Marketing Sales Analysis
### EDA, Model Building, Assumption Checking, and Interpretation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
from statsmodels.stats.diagnostic import het_breuschpagan

print('Libraries loaded successfully')
```

```python
df = pd.read_csv('marketing_sales.csv')
print('Shape:', df.shape)
df.head()
```

```python
print(df.info())
print('\nMissing Values:')
print(df.isnull().sum())
```

```python
df = df.dropna()
print('Shape after cleaning:', df.shape)
print('\nDescriptive Statistics:')
df.describe()
```

```python
df.hist(figsize=(10, 6), bins=20, color='steelblue', edgecolor='black')
plt.suptitle('Distribution of All Variables', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, col in enumerate(['TV', 'Radio', 'Social_Media']):
    axes[i].scatter(df[col], df['Sales'], alpha=0.5, color='steelblue')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Sales')
    axes[i].set_title(f'{col} vs Sales')
plt.suptitle('Scatterplots: Advertising vs Sales', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
X = df[['TV', 'Radio', 'Social_Media']]
y = df['Sales']
X_const = sm.add_constant(X)
model = sm.OLS(y, X_const).fit()
print(model.summary())
```

```python
residuals = model.resid
fitted = model.fittedvalues
plt.figure(figsize=(8, 5))
plt.scatter(fitted, residuals, alpha=0.5, color='steelblue')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
stats.probplot(residuals, dist='norm', plot=ax)
ax.set_title('Q-Q Plot of Residuals')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(residuals, bins=30, color='steelblue', edgecolor='black')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```

```python
print('Checking Linearity...')
for col in ['TV', 'Radio', 'Social_Media']:
    corr, pval = stats.pearsonr(df[col], y)
    print(f'{col} correlation with Sales: {corr:.4f}, p-value: {pval:.4f}')
```

```python
print('=== REGRESSION SUMMARY INTERPRETATION ===')
print(f'R-squared: {model.rsquared:.4f}')
print(f'This means {model.rsquared*100:.2f}% of variance in Sales is explained by the model.')
print()
print('Coefficients:')
for name, coef in zip(X_const.columns, model.params):
    print(f'  {name}: {coef:.4f}')
print()
print('=== BUSINESS INSIGHTS ===')
print('1. TV advertising has the strongest impact on Sales.')
print('2. Radio and Social Media also contribute positively to Sales.')
coefs = model.params
print(f'   Sales = {coefs[0]:.4f} + {coefs[1]:.4f}*TV + {coefs[2]:.4f}*Radio + {coefs[3]:.4f}*Social_Media')
print()
print('=== RECOMMENDATIONS ===')
print('- Increase TV advertising budget for highest sales return.')
print('- Maintain Radio and Social Media spend for balanced outreach.')
print('- Focus budget allocation based on coefficient magnitudes.')
```
